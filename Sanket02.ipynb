```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
%matplotlib inline
```

```python
df = pd.read_csv('train.csv', index_col = 'ID')
```

```python
df['Year Type'].replace(('Calendar','Fiscal'),(0,1),inplace=True)
```

```python
df.drop('Organization Group',axis=1,inplace=True)
```

```python
df.drop('Department',axis=1,inplace=True)
```

```python
df.drop('Union',axis=1,inplace=True)
```

```python
df.drop('Job Family', axis=1, inplace=True)
```

```python
df.drop('Job', axis=1, inplace=True)
```

```python
df.drop('Employee Identifier',axis=1,inplace=True)
```

```python
df = pd.get_dummies(df, columns=['Organization Group Code','Department Code','Union Code','Job Family Code','Job Code','Year'])
```

```python
df.head()
```

```python
test_df = pd.read_csv('test.csv',index_col = 'ID')
```

```python
test_df['Year Type'].replace(('Calendar','Fiscal'),(0,1),inplace=True)
```

```python
test_df.drop('Organization Group',axis=1,inplace=True)
test_df.drop('Department',axis=1,inplace=True)
test_df.drop('Union',axis=1,inplace=True)
test_df.drop('Job Family', axis=1, inplace=True)
test_df.drop('Job', axis=1, inplace=True)
test_df.drop('Employee Identifier',axis=1,inplace=True)
```

```python
test_df = pd.get_dummies(test_df, columns=['Organization Group Code','Department Code','Union Code','Job Family Code','Job Code','Year'])
```

```python
test_df.head()
```

```python
X = df.drop('Class', axis=1)
y = df['Class']
```

```python
X.head()
```

```python
for col in X.columns:
    if col not in test_df.columns:
        test_df[col] = 0
```

```python
for col in test_df.columns:
    if col not in X.columns:
        test_df.drop(col,axis=1,inplace=True)
```

```python
test_df.head()
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2 , random_state=5)
```

```python
from sklearn.ensemble import RandomForestClassifier
rf = RandomForestClassifier(n_estimators = 1000)
rf.fit(X_train, y_train)
```

```python
y_predict = rf.predict(X_test)
```

```python
from sklearn.metrics import confusion_matrix
```

```python
confusion_matrix(y_test, y_predict)
```

```python
final_df = pd.DataFrame()
for col in X_train.columns:
    final_df[col] = test_df[col]
```

```python
predict = rf.predict(final_df)
```
